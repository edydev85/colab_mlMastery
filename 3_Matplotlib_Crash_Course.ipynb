# Scatter Plot

```python
x = np.array([1,2,3])
y = np.array([2,4,6])
plt.scatter(x,y) # cria o grafico do tipo scatter
plt.xlabel("eixo x")
plt.ylabel("eixo y")
plt.show()
```
